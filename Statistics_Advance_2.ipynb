# Question 1: What is hypothesis testing in statistics?
# answer
Hypothesis testing in statistics is a formal procedure used to make decisions or inferences about a population based on sample data. It helps determine whether there is enough evidence in the data to support a specific claim or hypothesis about a population parameter (like mean, proportion, or variance).

# Question 2: What is the null hypothesis, and how does it differ from the alternative hypothesis?
# answer
**1. Null Hypothesis (H₀)**
The null hypothesis is the default assumption or statement that there is no effect, no difference, or no change in the population.
It is assumed true until evidence suggests otherwise.
Purpose: Serves as a baseline for testing.
**2. Alternative Hypothesis (H₁ or Ha)**
The alternative hypothesis is the statement that contradicts the null hypothesis.
It represents what you are trying to prove or detect.
It can be two-sided (not equal) or one-sided (greater than or less than).

# Question 3: Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.
# answer
The significance level, denoted by α, is the probability of making a Type I error in hypothesis testing.
A Type I error occurs when we reject the null hypothesis (H₀) even though it is true.
It represents the tolerance for risk of incorrect rejection.
**Common values of α:**
0.05 (5%) → most commonly used
0.01 (1%) → more stringent, lower chance of Type I error
0.10 (10%) → less strict, higher tolerance for error

# Question 4: What are Type I and Type II errors? Give examples of each.
# answer
**1Type I Error (False Positive)**
Definition: Occurs when we reject the null hypothesis (H₀) even though it is actually true.
In other words, we detect an effect that does not exist.
Probability of Type I error: Denoted by α (significance level).
# Example:
H₀: A new drug has no effect on patients.
Decision: You reject H₀ and claim the drug works.
Reality: The drug actually has no effect.
Result: Type I error (false positive).

**2. Type II Error (False Negative)**
Definition: Occurs when we fail to reject the null hypothesis (H₀) even though it is false.
In other words, we miss detecting a real effect.
Probability of Type II error: Denoted by β.
#Example:
h₀: A new drug has no effect on patients.
Decision: You fail to reject H₀ and say the drug does not work.
Reality: The drug actually works.
Result: Type II error (false negative).

# Question 5: What is the difference between a Z-test and a T-test? Explain when to use each.
# answer
# Z-Test
** Use when:**
The population standard deviation (σ) is known.
The sample size is large (typically n ≥ 30).
Example: Testing if the average height of adults differs from a known population mean, given population σ.
# T-Test
**use when:**
The population standard deviation (σ) is unknown.
Sample size is small (n < 30) or moderately large.
Example: Testing if the average marks of students in your class differ from 75, without knowing the population σ.

# Question 6: Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10       # Number of trials
p = 0.5      # Probability of success
size = 1000  # Number of experiments

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=np.arange(n+2)-0.5, edgecolor='black', density=True)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.xticks(range(n+1))
plt.show()
```

 # Question 7: Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results. sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6, 50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9]


```python
import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Population mean (under H0)
mu_0 = 50

# Sample mean and standard deviation
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation
n = len(sample_data)

# Calculate Z statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Calculate two-tailed p-value
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

# Display results
print(f"Sample Mean = {sample_mean:.3f}")
print(f"Sample Std Dev = {sample_std:.3f}")
print(f"Z-Statistic = {z_stat:.3f}")
print(f"P-Value = {p_value:.4f}")

# Decision at alpha = 0.05
alpha = 0.05
if p_value <= alpha:
    print("Reject H0: There is significant evidence that the mean is different from 50.")
else:
    print("Fail to reject H0: There is not enough evidence to say the mean is different from 50.")
```

# Question 8: Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Parameters
mu = 50         # Mean of the normal distribution
sigma = 5       # Standard deviation
n = 100         # Sample size
confidence = 0.95

# Simulate data
data = np.random.normal(mu, sigma, n)

# Sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
se = sample_std / np.sqrt(n)

# Calculate 95% confidence interval
ci = stats.t.interval(confidence, df=n-1, loc=sample_mean, scale=se)

print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci[0]:.2f}, {ci[1]:.2f})")

# Plot the data
plt.figure(figsize=(8,5))
plt.hist(data, bins=15, edgecolor='black', alpha=0.7)
plt.axvline(sample_mean, color='red', linestyle='--', label=f"Mean = {sample_mean:.2f}")
plt.axvline(ci[0], color='green', linestyle='--', label=f"95% CI Lower = {ci[0]:.2f}")
plt.axvline(ci[1], color='green', linestyle='--', label=f"95% CI Upper = {ci[1]:.2f}")
plt.title("Histogram of Simulated Normal Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

# Question 9: Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.


```python
import numpy as np
import matplotlib.pyplot as plt

# Sample dataset
data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
        50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
        50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
        50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Function to calculate Z-scores
def calculate_z_scores(dataset):
    mean = np.mean(dataset)
    std = np.std(dataset, ddof=1)  # sample standard deviation
    z_scores = [(x - mean) / std for x in dataset]
    return np.array(z_scores), mean, std

# Calculate Z-scores
z_scores, mean, std = calculate_z_scores(data)

# Display results
print(f"Mean of dataset = {mean:.2f}")
print(f"Standard Deviation of dataset = {std:.2f}")
print(f"Z-scores = {z_scores}")

# Plot histogram of Z-scores
plt.hist(z_scores, bins=10, edgecolor='black', alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.axvline(0, color='red', linestyle='--', label='Mean Z=0')
plt.legend()
plt.show()
```
